# Data cleaning

```python
# replace value in missing data => age
def clean_age(cols):
    age=cols[0]
    pclass=cols[1]

    #for i in range(Titanic.describe().loc['count'].max()):
    if pd.isnull(age):
        if pclass==1:
            return Titanic[Titanic['pclass']==1]['age'].mean()
        elif pclass==2:
            return Titanic[Titanic['pclass']==2]['age'].mean()
        elif pclass==3:
            return Titanic[Titanic['pclass']==3]['age'].mean()
    else:
        return age 
        
Titanic['age']=Titanic[['age','pclass']].apply(clean_age,axis=1)
```

```python
sns.heatmap(Titanic.isnull(),yticklabels=False,cbar=False,cmap="Set3")
```

```python
# delete a deck column
Titanic.drop('deck',axis=1,inplace=True)
sns.heatmap(Titanic.isnull(),yticklabels=False,cbar=False,cmap="Set3")
```

```python
# convert sex variable to numeric value
sex=pd.get_dummies(Titanic['sex'],drop_first=True)
print(sex)
```

```python
embarked=pd.get_dummies(Titanic['embarked'],drop_first=True)
print(embarked)
```

```python
Titanic.drop(columns=['sex','class','who','adult_male','embark_town','alive','alone','embarked'],axis=1,inplace=True)
```

```python
Titanic=pd.concat([Titanic,sex,embarked],axis=1)
```

# test and train split

```python
#seperate DATA to test and train
x=Titanic.drop('survived',axis=1)
y=Titanic['survived']

#seperate Data to Train and Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_trainStandard=scaler.fit_transform(X_train)
x_testStandard=scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
#create a object from Class
Logm=LogisticRegression()
Logm.fit(x_trainStandard,y_train)
```

```python
# test Model
prediction=Logm.predict(x_testStandard)
```

```python
# Evaluate Model
print(Logm.score(x_testStandard,y_test))

```

```python
#confusion Matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,prediction))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```
